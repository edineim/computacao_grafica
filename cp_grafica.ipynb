Importação das bibliotecas

```python
from PIL import Image, ImageChops
import math
import IPython.display as display
from google.colab import drive
drive.mount('/content/drive')
```

```python
# abre imagem base
base = Image.open('/content/drive/MyDrive/ComputacaoGrafica/mario-bros.jpg')

# obtendo informações sobre o tamanho da imagem
imgsize = base.size

innerColor = [0,0,0] #cor das laterais - preto
outerColor = [255,255,255] #cor do centro - branca
```

```python
# funcao de repetição para percorrer a imagem
for y in range(imgsize[1]):
    for x in range(imgsize[0]):

        # equacao da circunferencia
        # realizando o calculo do raio
        distanceToCenter = math.sqrt((x - imgsize[0]/2) ** 2 + (y - imgsize[1]/2) ** 2)

        # realizando o calculo do centro ate a borda
        distanceToEdge = float(distanceToCenter) / (imgsize[0]/2)

        # calculando valores r, g, b, correspondentes
        r = innerColor[0] * distanceToEdge + outerColor[0] * (1 - distanceToEdge)
        g = innerColor[1] * distanceToEdge + outerColor[1] * (1 - distanceToEdge)
        b = innerColor[2] * distanceToEdge + outerColor[2] * (1 - distanceToEdge)

        # pintando o pixel        
        base.putpixel((x, y), (int(r), int(g), int(b)))

# salvando a imagem gerada
base.save('/content/drive/MyDrive/ComputacaoGrafica/grad.jpg')

# apresentação da imagem gerada pelo gradiente
display.display(Image.open('/content/drive/MyDrive/ComputacaoGrafica/grad.jpg'))
```

```python
# carregando imagem base
base = Image.open('/content/drive/MyDrive/ComputacaoGrafica/mario-bros.jpg')
   
# carregando imagem com o gradiente
grad = Image.open('/content/drive/MyDrive/ComputacaoGrafica/grad.jpg')

# realizando a multiplicaçao das imagens
mux = ImageChops.multiply(base, grad)

# salvando a imagem resultante da multiplicação
mux.save('/content/drive/MyDrive/ComputacaoGrafica/mux.jpg')

# aprensentando imagem resultante da multiplicação
display.display(Image.open('/content/drive/MyDrive/ComputacaoGrafica/mux.jpg'))
```
